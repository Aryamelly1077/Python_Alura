Pedirle a una IA que nos genere 20 correos electrónicos con contenido aleatorio en strings (cadenas de texto).
En segundo lugar, crearemos una función que recibirá como argumento la lista de correos electrónicos y le pedirá a una IA que los resuma en una sola línea.

```python
# Instalamos groq en Colab
!pip install -q groq

# Importamos las librerias necesarios para que funcione nuestro codigo
import os
from google.colab import userdata
from groq import Groq

# Importamos la API KEY de groq que hemos almacenado en la seccion de secretos de Google Colab
userdata.get('groq_key')
os.environ['GROQ_API_KEY'] = userdata.get("groq_key")

# Creamos un cliente Groq
client = Groq()

#Realizamos la funcion indicada.
def resumir_correos(lista_correos):
  resumen = [];
  for num, correo in enumerate(lista_correos):
    respuesta = client.chat.completions.create(
    model="openai/gpt-oss-safeguard-20b", # Changed model to a currently supported one
    messages=[
        {
            "role": "user",
            "content": "Resume en una linea este correo por favor: "+ correo
        }
    ])
    texto = respuesta.choices[0].message.content
    print(f"Resumen del correo {num}: {texto}")
    resumen.append(texto)
  return resumen

#Listado de Correos
correos = [
    "Hola equipo,\n\nLes recuerdo que la reunión mensual de resultados se ha movido para mañana a las 10:00 AM. Por favor, asegúrense de tener listos sus reportes en la carpeta compartida antes de las 9:00 AM.\n\nSaludos cordiales,\nDirección de Operaciones.",
    "¡Hola! Tu pedido #84729 ya ha sido enviado y está en camino. Puedes realizar el seguimiento de tu paquete utilizando el enlace adjunto. Gracias por comprar con nosotros.\n\nAtentamente,\nEl equipo de Soporte al Cliente."
]

#Ejecutamos la funcion con el listado de correos previamente generado.
lista = resumir_correos(correos)

```
