Desafío:

1. Crear un archivo .txt a partir de una lista de preguntas provenientes de Python.
2. Leer esas preguntas desde el archivo .txt y guardarlas en una lista.
3. Obtener respuestas de un LLM para cada una.
4. Guardar los resultados en un nuevo archivo .csv.
5. Leer el archivo .csv usando Pandas.

```python
#Etapa #1

lista_de_preguntas = [
    "¿Cuál es la capital de Francia?",
    "¿en qué año se fundó Google?",
    "¿Cuál es el río más largo del mundo",
    "¿Cuál es el planeta más grande del sistema solar?",
    "¿Quién pintó la Mona Lisa?"
]

with open("lista_preguntas.txt", "w", encoding="utf-8") as archivo:
    for resumen in lista_de_preguntas:
        archivo.write(resumen + '\n')
```

```python
#Etapa #2
lista_preguntas = []
with open("lista_preguntas.txt", "r", encoding="utf-8") as archivo1:
    for linea in archivo1:
        lista_preguntas.append(linea.strip())

print(lista_preguntas)
```

```python
#Etapa #3

import os
from google.colab import userdata
from groq import Groq

userdata.get('groq_key')
os.environ['GROQ_API_KEY'] = userdata.get("groq_key")

client = Groq()

def contestar_preguntas(lista_preguntas):
  respuestas_lista = [];
  for num, pregunta in enumerate(lista_preguntas):
    respuesta = client.chat.completions.create(
    model="openai/gpt-oss-safeguard-20b",
    messages=[
        {
            "role": "user",
            "content": "Constesta la siguiente pregunta en una linea por favor: "+ pregunta
        }
    ])
    texto = respuesta.choices[0].message.content
    print(f"Respuesta {num}: {texto}")
    respuestas_lista.append({"pregunta": pregunta, "respuesta": texto})
  return respuestas_lista
```

```python
#Etapa #4
listado_respuestas = contestar_preguntas(lista_preguntas);
with open("respuestas_lista.csv", "w", encoding="utf-8") as archivo:
    for respon in listado_respuestas:
        archivo.write(f"{respon['pregunta']},{respon['respuesta'].replace(',',' ')}\n")
```

```python
#Etapa #5
import pandas as pd

df = pd.DataFrame(listado_respuestas)
df.to_csv("respuestas_lista.csv", index=False)

df
```
