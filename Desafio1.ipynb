```python
## Desafio Estrucuturas de datos
## Explicacion del desafio:
## De un listado de promedios queremos sumar un punto a cada promedio.
## Si el promedio pasa de 10 entonces dejar en 10.

# Iniciamos contador
n=0

#Lista de promedios brindada
promedios = [8.5, 6.5, 7.3, 9.5]

#Guardamos el tamaño de la lista en una variable
tamano = len(promedios)

#Estructura de Dato para recorrer lista mientras el contador sea menor que el tamaño de la lista
while n < tamano:
  print(f"promedio {n+1} antes: {promedios[n]}")
  if promedios[n] + 1.0 > 10:
    promedios[n] = 10
  else:
    promedios[n] = promedios[n]+1.0
  print(f"promedio {n+1} despues: {promedios[n]}")
  n = n + 1
print('Ya salimos del while')
```
